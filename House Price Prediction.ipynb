```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mlt
from sklearn import linear_model
```

```python
df = pd.read_csv("sample_data/california_housing_test.csv")
df.head(5)
```

```python
mlt.scatter(df.median_income,df.median_house_value,color ="red",marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['median_income']],df.median_house_value)
```

```python
reg.predict([[6.5]])
```
